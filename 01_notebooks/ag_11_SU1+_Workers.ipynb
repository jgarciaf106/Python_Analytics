# Data processing

Put here the core of the notebook. Feel free to further split this section into subsections.


```python
# filter data
data_filter = ["DIR", "EXEC", "MGR"]
worker_data_final = worker_data[worker_data["Management Level Category"].isin(data_filter)]

# fetch required columns
worker_data_final = worker_data_final[
    [
        "Preferred Name",
        "Email - Primary Work",
        "Management Level",
        "Hybrid L1 Code",
        "Pay Group Country Desc",
        "Metro Area",
        "Supervisor - Level 01 Preferred name",
        "Supervisor - Level 01 Email",
    ]
]

# split Full Names into separate colums
workers_name = worker_data_final["Preferred Name"].str.split(",", n=1, expand=True)
managers_name = worker_data_final["Supervisor - Level 01 Preferred name"].str.split(
    ",", n=1, expand=True
)

# add new colums to data frame
worker_data_final.insert(1, "Name", workers_name[1])
worker_data_final.insert(2, "Last Name", workers_name[0])
worker_data_final.insert(9, "Manager Name", managers_name[1])
worker_data_final.insert(10, "Manager Last Name", managers_name[0])

# trim white spaces
worker_data_final = ods.column_trim(worker_data_final)
```
