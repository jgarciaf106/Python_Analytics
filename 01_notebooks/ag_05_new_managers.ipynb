```python

# query statement
# report dates
hc_date = "'2021-12-31'" # this date should be last day of previous month
hire_date_from = "'2022-01-04'" # this date should be equal to the date for the requested report

# file name dates
report_date_from = "'2022-01-04'"
report_date_to = "'2022-02-03'"
queryPath = "C:\\Users\\garciand\\OneDrive - HP Inc\\Desktop\\Python_Analytics\\Python_Analysis\\Queries\\PD_New_Managers.sql"

with open(queryPath) as get:
    query = get.read()

# add query result to data frame
new_managers = ods.internal_query(query.format(hc_date, hire_date_from))

# export data to excel
nameFrom = dt.datetime.strptime(report_date_from.replace("'", ""), "%Y-%m-%d").strftime(
    "%b %d %Y"
)
nameTo = dt.datetime.strptime(report_date_to.replace("'", ""), "%Y-%m-%d").strftime(
    "%b %d %Y"
)
filename = "New Managers {0} - {1}".format(nameFrom, nameTo)
```
